```python
def dataprocessing(text):
    doc = nlp(text)
    tokens = []
    for token in doc:
        if token.is_stop:
            continue
        if token.is_punct:
            continue
        tokens.append(token.lemma_.lower())
    return " ".join(tokens)
    
        
        
```

```python
response = dataprocessing(data["text"][0])
```
