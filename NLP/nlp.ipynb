```python
import spacy
```

```python
nlp = spacy.load("en_core_web_trf")
```

```python
doc = nlp("Rav met Rose in New York on July 10, 2024, and they invested $5,000 in a startup called BrightTech. Later, Rose sent the details to rav.rose@gmail.com and asked Rav to check www.brighttech.com stock before 5 PM.")
```

```python
doc
```

```python
type(doc)
```

```python
for i in doc:
    print (i.text, i.is_alpha)
```

```python
for token in doc:
    if token.is_stop:
        print (token.text)
    
```

```python
for token in doc:
    if token.like_email:
        print (token.text)
```

```python
for token in doc:
    if token.like_url:
        print (token.text)
```

```python
for token in doc:
    print (token.text, token.pos_)
```

```python
for token in doc.ents:
    print (token.text, token.label_)
```

```python
print (nlp.pipe_names)
```

```python
for token in doc:
    print (token.text, token.lemma_)
```
